```python
import tkinter as tk
import time
import random


quiz_data = [
    {
        "question": "What is the primary cause of time dilation in special relativity?",
        "correct": "Relative motion close to the speed of light",
        "wrong": ["High temperatures", "Electric fields", "Magnetic interactions"]
    },
    {
        "question": "Which fundamental interaction is responsible for radioactive beta decay?",
        "correct": "Weak nuclear force",
        "wrong": ["Strong nuclear force", "Gravitational force", "Electromagnetic force"]
    },
    {
        "question": "What is a blackbody in physics?",
        "correct": "An idealized object that absorbs all incident radiation",
        "wrong": ["A body with zero temperature", "An object that reflects all light", "A black hole"]
    },
    {
        "question": "In general relativity, what replaces the force of gravity?",
        "correct": "Curvature of spacetime",
        "wrong": ["Gravitational waves", "Electromagnetic force", "Centripetal force"]
    },
    {
        "question": "What is the Pauli exclusion principle?",
        "correct": "No two fermions can occupy the same quantum state",
        "wrong": ["Electrons must repel each other", "Energy cannot be destroyed", "Only bosons can interact"]
    },
    {
        "question": "What is meant by quantum tunneling?",
        "correct": "Particles passing through energy barriers they classically shouldn't",
        "wrong": ["Quantum teleportation", "Wavefunction collapse", "Particle annihilation"]
    },
    {
        "question": "Which field theory unifies electromagnetic and weak forces?",
        "correct": "Electroweak theory",
        "wrong": ["Quantum chromodynamics", "Supersymmetry", "String theory"]
    },
    {
        "question": "What is the rest mass of a photon?",
        "correct": "Zero",
        "wrong": ["1.67 × 10^-27 kg", "9.11 × 10^-31 kg", "3 × 10^-30 kg"]
    },
    {
        "question": "Which of the following best describes a Bose-Einstein condensate?",
        "correct": "A state of matter at near absolute zero where particles occupy the same quantum state",
        "wrong": ["A plasma of high-energy ions", "A collection of fast-moving molecules", "An ionized gas at high temperature"]
    },
    {
        "question": "In particle physics, what is meant by 'color charge'?",
        "correct": "A property of quarks related to the strong interaction",
        "wrong": ["A type of electric charge", "The spin state of an electron", "The hue of visible photons"]
    }
]

class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Tkinter Flashcards")
        self.root.geometry("1000x1000")
        footer = tk.Frame(root, bg="gray", height=30)
        footer.pack(side="bottom", fill="x")

        self.question_label = tk.Label(root, text="", font=('Arial', 35), wraplength=350)
        self.question_label.pack(pady=20)

        self.buttons = []
        for i in range(4):
            btn = tk.Button(root, text="", font=('Arial', 35), width=100, command=lambda b=i: self.check_answer(b))
            btn.pack(pady=5)
            self.buttons.append(btn)

        self.result_label = tk.Label(root, text="", font=('Arial', 35))
        self.result_label.pack(pady=10)

        self.load_new_question()

        footer_label = tk.Label(footer, text="Made by CJC", bg="gray", fg="white")
        footer_label.pack(pady=5)

    def load_new_question(self):
        self.current = random.choice(quiz_data)
        all_answers = self.current["wrong"] + [self.current["correct"]]
        random.shuffle(all_answers)

        self.question_label.config(text=self.current["question"])
        for i in range(4):
            self.buttons[i].config(text=all_answers[i], state=tk.NORMAL)
        self.result_label.config(text="")

    def check_answer(self, button_index):
        for btn in self.buttons:
            btn.config(state=tk.DISABLED)

        chosen = self.buttons[button_index].cget("text")
        if chosen == self.current["correct"]:
            self.result_label.config(text="Right!", fg="green")
        else:
            self.result_label.config(text="Wrong!", fg="red")

        self.root.after(2000, self.load_new_question)


root = tk.Tk()
app = QuizApp(root)
root.mainloop()
```
